```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Initialize an empty list to store the dataframes
dataframes = []
nv_error_base = np.array([])
v_error_base = np.array([])
nv_error_vib = np.array([])
v_error_vib = np.array([])
nv_error_offset = np.array([])
v_error_offset = np.array([])
zero_vect = np.zeros(30)
p_num = 1
blocks = ['base', 'vib', 'offset']
# Loop through the range of numbers corresponding to your file names
for block in blocks:
    for i in range(7, 13):  # Assuming your files are numbered from 1 to 6
        # Construct the file name
        if i % 2 == 0:
            if block == 'base':
                file_name = f'../data/p{p_num}/p{p_num}_tf_{i}.csv'
                df = pd.read_csv(file_name)
                v_error_base = np.append(v_error_offset, df['error'].to_numpy())
                nv_error_base = np.append(nv_error_offset, zero_vect)
            elif block == 'vib':
                file_name = f'../data/p{p_num}/p{p_num}_tf_bv_{i}.csv'
                df = pd.read_csv(file_name)
                v_error_vib = np.append(v_error_offset, df['error'].to_numpy())
                nv_error_vib = np.append(nv_error_offset, zero_vect)
            elif block =='offset':
                file_name = f'../data/p{p_num}/p{p_num}_tf_of_{i}.csv'
                df = pd.read_csv(file_name)
                v_error_offset = np.append(v_error_offset, df['error'].to_numpy())
                nv_error_offset = np.append(nv_error_offset, zero_vect)
  
            df['Block_Number'] = i
            dataframes.append(df)
        else:
            if block == 'base':
                file_name = f'../data/p{p_num}/p{p_num}_nf_{i}.csv'
                df = pd.read_csv(file_name)
                nv_error_base = np.append(nv_error_offset, df['error'].to_numpy())
                v_error_base = np.append(v_error_offset, zero_vect)
            elif block == 'vib':
                file_name = f'../data/p{p_num}/p{p_num}_nf_bv_{i}.csv'
                df = pd.read_csv(file_name)
                nv_error_vib = np.append(nv_error_offset, df['error'].to_numpy())
                nv_error_vib = np.append(v_error_offset, zero_vect)
            elif block =='offset':
                file_name = f'../data/p{p_num}/p{p_num}_nf_of_{i}.csv'
                df = pd.read_csv(file_name)
                nv_error_offset = np.append(nv_error_offset, df['error'].to_numpy())
                v_error_offset = np.append(nv_error_offset, zero_vect)

            df['Block_Number'] = i
            dataframes.append(df)

combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
nv_error = np.ma.masked_equal(nv_error,0)
v_error = np.ma.masked_equal(v_error,0)

fig, ax = plt.subplots(figsize=[10,5])
ax.plot(nv_error, marker='.', linestyle='none', color='green', label='No Vision')
ax.plot(v_error, marker='.', linestyle='none', color='purple', label='Vision')
ax.axhline(0, linestyle='--')
for i in range(1,7):
    ax.axvline((30*i)-0.5, linestyle='--', color='black')
ax.set_title(f'Participant {p_num}: Alternating vision and no vision (biceps vibration)')
ax.set_ylabel(f'Target Error (\u00B0)')
ax.set_xlabel('Trial Number')
ax.spines['top'].set_visible(False)  # Remove top spine
ax.spines['right'].set_visible(False)  # Remove right spine
ax.legend(frameon=False)
ax.set_xticks([0, 30, 60, 90, 120, 150, 180])
ax.set_xticklabels([0, 30, 60, 90, 120, 150, 180])
fig.savefig(f'../results/p_{p_num}.png', dpi=1000)
```

```python
fig, ax, = plt.subplots(1,2, figsize=[12,6])
sns.pointplot(data=combined_df[combined_df['terminal_feedback']==0], x='trial_num', y='error', hue='Block_Number', 
              linestyles='none', ax=ax[0], palette='pastel')
ax[0].axhline(0, linestyle='--', color='black')
ax[0].set_title("No Feedback")
ax[0].set_ylim([-30,15])
handles, labels = ax[0].get_legend_handles_labels()
desired_handles = handles[0:3] 
desired_labels = ['Block 1', 'Block 2', 'Block 3']
# desired_labels = labels[1:4]  # Keep only the first three labels
ax[0].legend(desired_handles, desired_labels, frameon=False)

sns.pointplot(data=combined_df[combined_df['terminal_feedback']==1], x='trial_num', y='error', hue='Block_Number', 
              linestyles='none', ax=ax[1], palette='pastel')
ax[1].axhline(0, linestyle='--', color='black')
ax[1].set_title("Terminal Feedback")
ax[1].set_ylim([-30,15])
plt.legend().set_visible(False)


sns.despine()
plt.show()
```
